```python
import pandas as pd
import json

with open('data.json') as f:
    data = json.load(f)

rows = []
for city, pollutants in data.items():
    for pollutant, values in pollutants.items():
        if isinstance(values, int):
            # Handle the case where the value is an integer
            rows.append({
                'city': city,
                'pollutant': pollutant,
                'concentration': values,
                'aqi': None  # Set AQI to None since there is no value
            })
        else:
            # Handle the case where the value is a dictionary
            rows.append({
                'city': city,
                'pollutant': pollutant,
                'concentration': values['concentration'],
                'aqi': values['aqi']
            })

df = pd.DataFrame(rows)
df.to_csv('data.csv', index=False)
```

```python
#Y IS DEPENDENT VARIABLE.X IS INDEPENDENT VARIABLES.
from sklearn.model_selection import train_test_split
X=df8.drop("Effect",axis=1)
Y=df8["Effect"]
print("SHAPE OF X:",X.shape)
print("SHAPE OF Y:",Y.shape)


#80 percent data is used for training while 20 is used for testing.
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
print("SHAPE OF X_train:",X_train.shape)
print("SHAPE OF Y_train:",Y_train.shape)
print("SHAPE OF X_test:",X_test.shape)
print("SHAPE OF Y_test:",Y_test.shape)
```

```python
# FEATURE SCALING USING STANDARD SCALER

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
scaler.fit(X_test)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# MODEL IMPLEMENTATION

**RANDOM FOREST CLASSIFIER**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier()

# fit the model with the training data
model.fit(X_train,Y_train)

model.n_estimators=100
# number of trees used
print('Number of Trees used : ', model.n_estimators)


```

**MODEL EVALUATION**

```python
# predict the target on the train dataset
predict_train = model.predict(X_train)
 

# Accuray Score on train dataset
accuracy_train = accuracy_score(Y_train,predict_train)
print('Accuracy_score on train dataset : ', (accuracy_train*100))

# predict the target on the test dataset
predict_test = model.predict(X_test)
 

# Accuracy Score on test dataset
accuracy_test = accuracy_score(Y_test,predict_test)
print('Accuracy_score on test dataset : ',(accuracy_test*100))

```

```python
from sklearn.metrics import r2_score
predict_test1=model.predict(X_test)
rRegressorM1 = r2_score(Y_test,predict_test1)   #checking accuracy of predicted values
print("R Squared value=",rRegressorM1)
```

R Squared Value close to 1,which means it explains all variations perfectly.
